```python
# task 2 (group : Daniel & Putri)
# question:
# if n is off, print "tidak normal"
# if n is even and in the inclusive range [10-100] to, print Normal
# if n is even and in the inclusive range [10-200] to, print Tidak Normal
# if n is even and greater than 200, print Normal

N = int(input("Nilai N:"))

if (N % 2 == 1):
  print("Tidak Normal")
elif (N >= 10 and N <= 100):
  print("Normal")
elif (N >= 10 and N <= 200):
  print("Normal")
elif (N > 200):
  print("Normal")
else:
  print("...")
```
